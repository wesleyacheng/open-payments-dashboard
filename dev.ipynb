# Pre-Process Data

```python
# setup final dataframe
dff = pd.DataFrame()
```

```python
# get name of doctor
dff_name = df[["Physician_Profile_ID", "Physician_First_Name", "Physician_Middle_Name", "Physician_Last_Name", "Physician_Name_Suffix"]]

# drop duplicates
dff_name.drop_duplicates("Physician_Profile_ID", inplace=True)

# merge name into one column
def make_doctor_name(row):
    doctor_name_list = []
    if not pd.isna(row["Physician_First_Name"]):
        doctor_name_list.append(row["Physician_First_Name"].strip("./").upper())
    if not pd.isna(row["Physician_Middle_Name"]):
        doctor_name_list.append(row["Physician_Middle_Name"].strip("./").upper())
    if not pd.isna(row["Physician_Last_Name"]):
        doctor_name_list.append(row["Physician_Last_Name"].strip("./").upper())
    if not pd.isna(row["Physician_Name_Suffix"]):
        doctor_name_list.append(row["Physician_Name_Suffix"].strip("./").upper())
    return " ".join(doctor_name_list)
dff_name["DOCTOR_NAME"] = dff_name.apply(lambda row: make_doctor_name(row), axis=1)

# rename columns
dff_name = dff_name[["Physician_Profile_ID", "DOCTOR_NAME"]]
dff_name.rename(
    columns={
        "Physician_Profile_ID":"DOCTOR_ID",
    },
    inplace=True
)

dff_name.head()
```

```python
# merge payment data with name
dff = dff_name
dff.head()
```

```python
# get address of doctor
address_cols = [
    "Physician_Profile_ID",
    "Recipient_Primary_Business_Street_Address_Line1",
    "Recipient_Primary_Business_Street_Address_Line2",
    "Recipient_City",
    "Recipient_State",
    "Recipient_Zip_Code",
    "Recipient_Country",
    "Recipient_Province",
    "Recipient_Postal_Code"
]
dff_address = df[address_cols]

# drop duplicates
dff_address.drop_duplicates("Physician_Profile_ID", inplace=True)

# merge address into one column
def make_address_name(row):
    address_name_list = []
    for col in address_cols[1:]:
        if not pd.isna(row[col]):
            address_name_list.append(str(row[col]).strip("./").upper())
    return ", ".join(address_name_list)
dff_address["DOCTOR_ADDRESS"] = dff_address.apply(lambda row: make_address_name(row), axis=1)

# rename columns
dff_address = dff_address[["Physician_Profile_ID", "DOCTOR_ADDRESS"]]
dff_address.rename(
    columns={
        "Physician_Profile_ID":"DOCTOR_ID",
    },
    inplace=True
)

dff_address.head()
```

```python
# merge address data with existing df
dff = pd.merge(dff, dff_address, left_on="DOCTOR_ID", right_on="DOCTOR_ID")
dff.head()
```

```python
# get total payments made to same physician
dff_payments = df.groupby("Physician_Profile_ID")["Total_Amount_of_Payment_USDollars"].agg("sum").to_frame().reset_index()
dff_payments.rename(
    columns = {
        "Physician_Profile_ID":"DOCTOR_ID",
        "Total_Amount_of_Payment_USDollars":"TOTAL_PAYMENTS",
    },
    inplace=True
)
dff_payments.head()
```

```python
# merge payments data with existing df
dff = pd.merge(dff, dff_payments, left_on="DOCTOR_ID", right_on="DOCTOR_ID")
dff.head()
```

```python
# change doctor_id from float to int
dff["DOCTOR_ID"] = dff.apply(lambda row: int(row["DOCTOR_ID"]), axis=1)
dff.head()
```

```python
dff.info()
```

```python
# set doctor id as index
dff.set_index("DOCTOR_ID", inplace=True)
dff.head()
```

```python
# sort with highest total payments
dff.sort_values("TOTAL_PAYMENTS", ascending=False, inplace=True)
dff.head()
```

```python
dff.reset_index().head()
```

```python
# save to file for dashboard
dff.reset_index().to_csv("2019_general_payments.csv", index=False)
```
